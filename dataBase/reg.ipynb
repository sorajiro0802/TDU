```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
%precision 3
%matplotlib inline
```

```python
df = pd.read_excel("data/2019.xlsx", sheet_name="2022data")
```

```python
direction = ["Hori", "Vert", "RU", "RD"]
section = [f"S{i}" for i in range(1, 10)]
objectiveVariable = "開放的"
for d in direction:
    for s in section:
        formula = f"{objectiveVariable} ~ {d}_{s}"
        res = smf.ols(formula, df).fit()
        print(f"{res.rsquared  :.6f}")
    print("\t")
```

```python
formula = "広い ~ 明度平均 + 水平"
result = smf.ols(formula, df).fit()
result.summary()
```

```python
width = np.array(df["広い"])
brAve = np.array(df["明度平均"])
hori = np.array(df["水平"])
n = len(width)
```

```python
width_norm = (width - np.average(width))/ np.sqrt(np.var(width, ddof=1))
brAve_norm = (brAve - np.average(brAve))/ np.sqrt(np.var(brAve, ddof=1))
hori_norm  = (hori  - np.average(hori)) / np.sqrt(np.var(hori,  ddof=1))
```

```python
norm_df = pd.DataFrame({"width_norm": width_norm,
                        "brAve_norm": brAve_norm,
                        "hori_norm" : hori_norm} ,
                        index=["sample{}.jpg".format(i) for i in range(1, n+1)])
```

```python
norm_formula = "width_norm ~ brAve_norm + hori_norm"
norm_result = smf.ols(norm_formula, norm_df).fit()
norm_result.summary()
```
